```python
df.head()
```

```python
mu_2006 = 93.29
```

```python
x_bar_2012       = df_sample.time.mean()
sigma_x_bar_2012 = df_sample.time.std()
stderr_2012      = sigma_x_bar_2012/sqrt(len(df_sample))
x_2012_est       = x_bar_2012 + stderr_2012*np.array([-1.96, 1.96])
x_2012_est
```

```python
from interval import Interval

# if mu_2006 lies within interval, then we fail to reject H0
if mu_2006 in Interval(x_2012_est[0], x_2012_est[1]):
    print("Fail to reject H0 at 95% CL")
else:
    print("Reject H0 at 95% CL")
```

```python
# ex 4.21

# H0: ages between 2006 and 2012 are the same
# H1: ages between 2006 and 2012 have changed

# 2006 mean is given
mu_2006 = 36.13

# 2012 is estimated from 100 samples
N                = len(df_sample)
x_bar_2012       = df_sample.age.mean()
sigma_x_bar_2012 = df_sample.age.std()
stderr_2012      = sigma_x_bar_2012/sqrt(N)
x_2012_est       = x_bar_2012 + stderr_2012*np.array([-1.96, 1.96])

# test
if mu_2006 in Interval(x_2012_est[0], x_2012_est[1]):
    print("Fail to reject H0 at 95% CL")
else:
    print("Reject H0 at 95% CL")
```

```python
display(Latex(r"$\mu_{2006}     = %1.2f$"          % mu_2006))
display(Latex(r"$\bar{x}_{2012} = (%1.2f, %1.2f)$" % (x_2012_est[0],x_2012_est[1])))
```

```python
# ex 4.22

# H0: mu = 650
# H1: mu != 650

# mu
mu = 650

# data
df      = pd.read_table('data/housing.txt')
n       = len(df)
xbar    = df.cost.mean()
se_xbar = df.cost.std()/sqrt(n)
x_est   = Interval(xbar -1.96*se_xbar, xbar + 1.96*se_xbar)

# test
if mu in x_est:
    print("Fail to reject H0 at 95% CL")
else:
    print("Reject H0 at 95% CL")
    
x_est    
```

```python
# section 4.3.4

# H0: mu = 7
# H1: Mu > 7

alpha      = 0.05
mu         = 7
n          = 110
xbar       = 7.42
sigma_xbar = 1.75
se_xbar    = sigma_xbar/sqrt(n)

# ex 4.29
se_xbar
```

```python
# test
x_est = Interval(xbar -1.96*se_xbar, xbar + 1.96*se_xbar)
if mu in x_est:
    print("Fail to reject H0 at 95% CL")
else:
    print("Reject H0 at 95% CL")
    
x_est   
```

```python
z_xbar = (xbar - mu)/se_xbar
z_xbar
```

```python
p_value = 1 - scipy.stats.norm.cdf(z_xbar) # right sided
p_value
```

```python
def hyp_test(z, alpha=0.05, two_sided=False):
    z = -1.0*abs(z)
    p = scipy.stats.norm.cdf(z)
    if (two_sided):
        p = 2*p
    cl = 100*(1.0-alpha)
    if (p < alpha):    
        print("Reject H0 at %1.0f%% CL with p = %1.2e" % (cl, p))
    else:
        print("Fail to reject H0 at %1.0f%% CL with p = %1.2e" % (cl, p))
    
hyp_test(z_xbar, alpha)
```

```python
# ex 4.31
hyp_test(z_xbar, alpha=0.01 )
hyp_test(z_xbar, alpha=0.001)
```

```python
# ex 4.32-4.34

# H0: mu = 46.99
# H1: mu < 46.99

alpha      = 0.01
mu         = 46.99
n          = 52
xbar       = 44.17
sigma_xbar = 4.15
se_xbar    = sigma_xbar/sqrt(n)
z_xbar     = (xbar - mu)/se_xbar
p_value    = scipy.stats.norm.cdf(z_xbar) # left sided
hyp_test(z_xbar, alpha)
p_value
```

```python
# ex 4.6

# H0: mu = 7
# H1: Mu != 7
# two sided test

alpha      = 0.05
mu         = 7
n          = 72
xbar       = 6.83
sigma_xbar = 1.8
se_xbar    = sigma_xbar/sqrt(n)
z_xbar     = (xbar - mu)/se_xbar

hyp_test(z_xbar, alpha, two_sided=True)
```

### 4.4 Examining the central limit theorem

```python
def comparison_plot(X, M=1e6, N=2, xlim=[-1,1], ax=plt):
    # generate data
    np.random.seed(1)
    m = X.rvs((M,N))
    
    # sampling distibution of mean
    mu = m.mean(axis=1)
    
    # point estimate of sampling distibutions of mean: xbar
    xbar    = mu.mean()
    se_xbar = mu.std()
    
    # display
    ax.hist(mu, histtype='step', normed=True, bins=np.linspace(-10, 10, 1000))
    ax.plot(x, scipy.stats.norm(loc=xbar, scale=se_xbar).pdf(x), 'r--')
    ax.set_xlim(xlim)
    (mu.mean(), xbar, se_xbar)
```

```python
# reproduce figure 4.20

M          = 1e6
x          = np.linspace(-10,10,1000)
rv_uniform = scipy.stats.uniform()
rv_exp     = scipy.stats.expon()
rv_lognorm = scipy.stats.lognorm(1.3)
fig, axes  = plt.subplots(5, 3, figsize=(20,10))

# population dist
axes[0,0].plot(x, rv_uniform.pdf(x)); axes[0,0].axis([-1,2,0,1.5]); axes[0,0].text(0.35, 1.25, 'uniform'   )
axes[0,1].plot(x, rv_exp.pdf(x)    ); axes[0,1].axis([-2,6,0,1.5]); axes[0,1].text(0.35, 1.25, 'exp'       )
axes[0,2].plot(x, rv_lognorm.pdf(x)); axes[0,2].axis([-2,6,0,1.0]); axes[0,2].text(0.35, 0.85, 'log-normal')

# sampling dist and estimates
comparison_plot(rv_uniform, M=M, N=2 , xlim=[-1,2], ax=axes[1,0]); axes[1,0].text(-0.5, 1.5, ("n = %d" % 2))
comparison_plot(rv_exp    , M=M, N=2 , xlim=[-1,3], ax=axes[1,1])
comparison_plot(rv_lognorm, M=M, N=2 , xlim=[-6,9], ax=axes[1,2])
comparison_plot(rv_uniform, M=M, N=5 , xlim=[-1,2], ax=axes[2,0]); axes[2,0].text(-0.5, 1.5, ("n = %d" % 5))
comparison_plot(rv_exp    , M=M, N=5 , xlim=[-1,3], ax=axes[2,1])
comparison_plot(rv_lognorm, M=M, N=5 , xlim=[-3,9], ax=axes[2,2])
comparison_plot(rv_uniform, M=M, N=12, xlim=[-1,2], ax=axes[3,0]); axes[3,0].text(-0.5, 1.5, ("n = %d" % 12))
comparison_plot(rv_exp    , M=M, N=12, xlim=[-1,3], ax=axes[3,1])
comparison_plot(rv_lognorm, M=M, N=12, xlim=[-1,6], ax=axes[3,2])
comparison_plot(rv_uniform, M=M, N=30, xlim=[-1,2], ax=axes[4,0]); axes[4,0].text(-0.5, 1.5, ("n = %d" % 30))
comparison_plot(rv_exp    , M=M, N=30, xlim=[-1,3], ax=axes[4,1])
comparison_plot(rv_lognorm, M=M, N=30, xlim=[-1,6], ax=axes[4,2])

# display
#fig.savefig("fig4p20.pdf")
fig.show()
```
