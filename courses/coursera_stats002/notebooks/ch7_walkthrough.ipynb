Linear regression attempts to find a linear relationship between x, the *predictor* variable, to y, the *response* variable.  This relations ship is of the form:

\begin{align}
    y = \beta_0 + \beta_1 x
\end{align}

where $\beta_i$ are the parameters of this linear model.  The point estimates are determined by data and are written as $b_i$.

#### residuals

The residuals are the *leftover* variation in the data after account for the fit in your linear model.  More specifically:

\begin{align}
    e_i &= y_i - \hat{y}_i \\
        &= y_i - f(x_i) \\
        &= y_i - \left( \beta_0 + \beta_1 x_i \right)
\end{align}

#### ex 7.3

```python
def f(x, b0=41, b1=0.59):
    return (b0 + b1*x)
```

#### correlation

Correlation, which always takes values between -1 and 1, describes the strength of the **linear relationship** between two variables.  We denote the correlation by $R$.

We can compute the correlation between observations $(x_i, y_i)$ by using the formula

\begin{align}
    R = \frac{1}{n-1} \sum_{i=1}^{n} \frac{x_i - \bar{x}}{s_{x}} \frac{y_i - \bar{y}}{s_{y}}
\end{align}

```python
xbar = x.mean()
ybar = y.mean()
s_x  = x.std()
s_y  = y.std()
n    = len(x)
R    = (1.0/(n - 1.0)) * np.sum(((x - xbar)/s_x) * ((y - ybar)/s_y))
R
```

### Fitting a line by least squares regression

In order to fit the line, we want to minimize the squared sum of the residuals:

\begin{align}
    \sum_{i=1}^{n} e_i^2
\end{align}

Using squared residuals ensures that large residuals are weighted more in penalizing the fit.

When fitting a least squares line, we generally require

* Linearity: data should be someone linear to expect a good fit
* Nearly normal residuals: no major outliers and a normal model for the uncertainty can be assumed.
* Constant variablity: the fluctuation of the residuals is constant over the whole data

#### fit by hand

* The slope of the least squares fit is estimated by $b_1 = \frac{s_y}{s_x} R$
* The point $(\bar{x}, \bar{y})$ is a point on the fit line.

Then you can write this in point-slope form:

\begin{align}
    y - \bar{y} = b_1 \left(x - \bar{x} \right)
\end{align}

Simplifying into slope-intercept form:

\begin{align}
    y = b_0 + b_1 x
\end{align}

where

\begin{align}
    b_1 &= \frac{s_y}{s_x} R \\
    b_0 &= \bar{y} - b_1 \bar{x}
\end{align}



```python
# fit by hand
xbar = x.mean()
ybar = y.mean()
s_x  = x.std()
s_y  = y.std()
n    = len(x)
R    = (1.0/(n - 1.0)) * np.sum(((x - xbar)/s_x) * ((y - ybar)/s_y))
b1   = s_y * R / s_x
b0   = ybar - b1*xbar
(b1, b0, R)
```

```python
yhat2 = b0 + b1*xi
plt.plot(xi, yhat, 'r--')
plt.scatter(x, y)
plt.xlabel('family income ($1k)'           , fontsize=20)
plt.ylabel('Gift Aid From University ($1k)', fontsize=20)
plt.xlim([0, 300])
plt.plot(xi, yhat , 'r-' , label='polyfit', linewidth=4)
plt.plot(xi, yhat2, 'g--', label='by hand', linewidth=4)
plt.legend()
plt.show()
```

```python
# get the uncertainties of the fit
b, v_b = polyfit(x=x, y=y, deg=1, cov=True)
(b, v_b)
```

#### Using $R^2$

It is useful to use the square of the correlation to explain the linearity of the data.  It turns out that you can calculate $R^2$ using

\begin{equation}
    R^2 = \frac{s_y^2 - s_{e}^2}{s_y^2}
\end{equation}

### 7.3 Types of outliers in linear regression

**Leverage:** Points that fall away from the center of the cloud of data tend to pull harder on the line, so we call them points with *high leverage*.

**CAUTION:** Don't ignore outliers when fitting the data.  Often the outliers are the interesting pieces.

### 7.4 Inference for linear regression

here we show how to extract the uncertainties on the fit parameters $b_i$.

#### electron example

Here we determine if the higher the unemployment rate, the worse the president's party does in the midtem elections.

#### figure 7.20 

Look at the raw data before we fit.

```python
df = pd.read_table('data/unemployPres.txt')
df = df.ix[~np.in1d(df.year, [1935, 1939])]
df.head()
```

```python
# masks 
rep = (df.party == "Republican")
dem = (df.party == "Democratic")
```

```python
# raw plot
plt.scatter(df.ix[rep].unemp, df.ix[rep].change, c='r', marker='^', label = "rep", s=80, edgecolors='none')
plt.scatter(df.ix[dem].unemp, df.ix[dem].change, c='b', marker='o', label = "dem", s=80, edgecolors='none')
plt.xlabel('percent unemployed', fontsize=15)
plt.ylabel('percent in seats for potus\' party', fontsize=15)
plt.legend()
plt.show()
```

The relationship can be fit with a linear plot.  We can use this linear fit to classify relationship and determine if there is a relationship.

* H0: $\beta_1 = 0$
* H1: $\beta_1 \lt 0$

```python
x = df.unemp
y = df.change
b, v_b = polyfit(x=x, y=y, deg=1, cov=True)
(b, v_b)
```

```python
e_b = sqrt(np.diag(v_b))
(b, e_b)
```

```python
display(Latex(r"$\hat{\beta}_1 = %1.3f \pm %1.3f$" % (b[0], e_b[0])))
```

```python
# model
xi   = np.linspace(0,14,100)
yhat = xi*b[0] + b[1]

# raw plot
plt.scatter(df.ix[rep].unemp, df.ix[rep].change, c='r', marker='^', label = "rep", s=80, edgecolors='none')
plt.scatter(df.ix[dem].unemp, df.ix[dem].change, c='b', marker='o', label = "dem", s=80, edgecolors='none')
plt.plot(xi, yhat, 'r--', label="linear fit")
plt.xlabel('percent unemployed', fontsize=15)
plt.ylabel('percent in seats for potus\' party', fontsize=15)
plt.xlim([0,14])
plt.legend()
plt.show()
```
