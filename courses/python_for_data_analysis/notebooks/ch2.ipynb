### Counting Time Zones in Pure Python

```python
# will not work --> not all records have the 'tz' field
time_zones = [r['tz'] for r in records]
```

```python
time_zones = [r['tz'] for r in records if 'tz' in r]
```

```python
# print first 10 -- leave the unknown in there
time_zones[:10]
```

```python
# func to count time zones
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
counts = get_counts(time_zones)
print counts
```

```python
# alternative get_counts implementation
def get_counts2(sequence):
    counts = defaultdict(int) # values will initialize to 0
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
counts = get_counts(time_zones)
print counts
```

```python
# find the top 10 time zones and their counts
def top_counts(count_dict, n=10):
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    value_key_pairs.sort()
    return value_key_pairs[-n:]
```

```python
top_counts(counts)
```

### Measuring Rating Disagreements

```python
# add a column called 'diff'
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']                                                    
```

```python
sorted_by_diff = mean_ratings.sort_index(by='diff')
```

```python
# F prefered
sorted_by_diff[:15]
```

```python
# male prefered movies that females didn't rate
sorted_by_diff[::-1][:15]
```

```python
# find movies with most disagreement between viewers
# i.e. find the movies with the most "spread" --> variance

# get the std dev grouped by title
rating_std_by_title = data.groupby('title')['rating'].std()

# filter down to active titles 
rating_std_by_title = rating_std_by_title.ix[active_titles]

# order by value in descending order
rating_std_by_title.order(ascending=False)[:10]
```

```python
# load all the data into a single DataFrame

# 2013 is the last available year
years = range(1880, 2014) # one past last year

# an array of DataFrames to be filled
pieces = []
columns = ['name', 'sex', 'births']
for year in years:
    path = 'data\\ch2\\names\\yob%d.txt' % year
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)
    
# combine them into one DataFrame
names = pd.concat(pieces, ignore_index=True)
```

```python
names.info()
```

```python
names[:10]
```

### plot total births

```python
total_births = names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
```

```python
total_births.tail()
```

```python
total_births.plot(title='Total births by sex and year')
```

```python
# assign a name probability
def add_prob(group):
    # integer division floors
    births = group.births.astype(float)
    group['prob'] = births/births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prob)
```

```python
names.info()
```

```python
# test probabilities sum to one
print "test =", np.allclose(names.groupby(['year', 'sex']).prob.sum(), 1)
```

```python
names[:10]
```

### The "Last Letter" Revolution

```python
# extract last letter from name column
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name='last_letter'
```

```python
table = names.pivot_table('births', rows=last_letters, cols=['sex', 'year'], aggfunc=sum)
```

```python
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prob = subtable / subtable.sum().astype(float)
```

```python
# plot result
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 1, figsize=(10, 8))
letter_prob['M'].plot(kind='bar', rot=0, ax=axes[0], title="Male")
letter_prob['F'].plot(kind='bar', rot=0, ax=axes[1], title="Female", legend=False)
```
