### 2.1 Sampling Theorem

The key bridge between the analog and the digital world is *sampling*.  

In terms of electroncis:

* an analog signal voltage is applied to a sample-and-hold circuit
* it is held there by a tiny bit of capacitance 
* a bank of compararators marks off the proportional integer value
* these values are the **digital sample** of the analog signal.

The sampling theorem provides the mathematical mechanism that models this.

From [Wikipedia](http://en.wikipedia.org/wiki/Nyquist%E2%80%93Shannon_sampling_theorem), 

> If a function x(t) contains no frequencies higher than B hertz, it is completely determined by giving its ordinates at a series of points spaced 1/(2B) seconds apart.

Recall that we can represent an analog signal using a Fourier Series expansion:

\begin{align}
    x(t) = \sum_{n} a_n e^{-j \omega_n t}
\end{align}

where the coefficents $a_n$ are given by

\begin{align}
    a_n = \frac{1}{T} \int_0^T dt \ x(t) \ e^{-j \omega_n t}
\end{align}

Here, in order to generate the descrete points $a_n$, which uniquely describe the signal, we must integrate over the *entire* function x(t).  We avoid this by using sampling.

Suppose we are given a set of samples from x(t), $[x_1, x_2, ..., x_n]$ and we are asked to reconstruct the original function.  How do we do this?  

Fist, we attempt to use a straight line connection between the points.

### 2.2 Reconstruction

Let's consider another way of examining the signal.

Condider a band limited signal: $X(f) = 0$ where $|f| \gt W$ and $X$ is the Fourier transform (FT) of $x(t)$, which we'll denote as $\mathscr{F}(x)$.

The signal can be reproduced with the inverse FT:

\begin{equation}
    x(t) = \int_{-W}^{W} df\ X(f)\ e^{+j 2 \pi f t}
\end{equation}

We can pretend $X(f)$ is periodic in $2W$ and expand it as a Fourier Series (FS):

\begin{align}
    X(f) &= \sum_k a_k\ e^{\frac{-j 2 \pi k f}{2W}} \\
    a_k  &= \frac{1}{2W} \int_{-W}^{W} df\ X(f)\ e^{\frac{+j 2 \pi k f}{2W}}
\end{align}

These coefficients looks strikingly similar to the full inverse FT shown above.  In, fact you can detemine $a_k$ from $x(t)$ by lining up the terms:

\begin{align}
    a_k = \frac{1}{2W} x \left(t = \frac{k}{2W} \right)
\end{align}

To see how this is useful, subsitute the $a_k$ and $X(f)$ definitions above to get

\begin{align}
    x(t) &= \int_{-W}^{W} df \sum_k \frac{1}{2W} x \left(t = \frac{k}{2W} \right) e^{\frac{-j 2 \pi k f}{2W}}\ e^{+j 2 \pi f t} \\
         &= \sum_k \frac{1}{2W} x \left(t = \frac{k}{2W} \right) \int_{-W}^{W} df\ e^{\frac{-j 2 \pi k f}{2W}+j 2 \pi f t}\\
         &= \sum_k \frac{1}{2W} x \left(t = \frac{k}{2W} \right) {\rm sinc}(k - 2tW)
\end{align}

where, 

\begin{align}
    {\rm sinc}(x) = \frac{{\rm sin}(\pi x)}{\pi x}
\end{align}

This is exactly what we wanted -- a formula to reproduce the $x(t)$ from its samples.

Since our samples are spaced $t = k/f_s$, we'll use $W = f_s/s$.  The following is a repeat visualization of the squared errors using this new model.

```python
# listing 2.6
t  = np.linspace(-1,1,100)
ts = np.arange(-1, 1+1/fs, 1/fs)
n  = len(ts)
sm = 0
for k in np.arange(-n, n):
    sm += x_func(k/fs, f)*sinc(k - fs*t)
     
fig,ax=subplots()    
ax.plot(t, sm, '--', t, x_func(t, f), ts, x_func(ts, f),'o')
ax.set_title('Sampling Rate=%3.2f Hz' % fs, fontsize=18 )
plt.show()
```

```python
# listing 2.6
fig,ax1=subplots()

# shade the area to represent the difference
ax1.fill_between(t, sm, x_func(t, f), facecolor='red')
ax1.set_xlabel('time ($Hz$)'               , fontsize=18)
ax1.set_ylabel('amplitude of $x(t)$ ($Hz$)', fontsize=18)

# compute the squared errors and display on same figure
ax2 = ax1.twinx() # clone
e2  = (sm - x_func(t, f))**2
ax2.plot(t, e2, 'g--')
#ax2.axis(xmin=0, ymax=e2.max())
ax2.set_ylabel('squared error, (${Hz}^2$)', fontsize=18)

ax1.set_title('Errors with linear interploation', fontsize=18)
plt.show()
```

The squared-error is much better here than what we obtained with a linear plot.  This sort of interpolation using the *sinc* function is called the *Whittaker* inerpolating function.  

```python
k  = 0 # pick a point
fs = 2 # makes easier to read

# plot
fig, ax = subplots()
ax.plot(t, np.sinc(k   - fs*t),
        t, np.sinc(k+1 - fs*t), '--', k/fs, 1, 'o', (+k)/fs, 0, 'o',
        t, np.sinc(k+1 + fs*t), '--', k/fs, 1, 'o', (-k)/fs, 0, 'o')

# label
ax.hlines(0, -1.0, 1.0) # horizontal line
ax.vlines(0, -0.2, 1.0) # horizontal line
ax.annotate('sample values go here',
            xy=(0,1),             # arrowhead position
            xytext=(-1+0.1, 1.1), # text position
            arrowprops={'facecolor':'red', 'shrink':0.05})
ax.annotate('no interference here',
            xy=(0,0),             # arrowhead position
            xytext=(-1+0.1, 0.5), # text position
            arrowprops={'facecolor':'green', 'shrink':0.05})

plt.plot()
```

Notice that in the above plot for *sinc* function has the corresponding current point plus the points before and after.  Only the current point has a non-zero weight at the point of interest. This essentially acts a our $\delta$ function in that the other points contribute nothing and only the *sinc* associated with the current term has any weight.
