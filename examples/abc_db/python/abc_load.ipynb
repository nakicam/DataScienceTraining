## Load some teams

```python
team1_pic  = plt.imread('images/team1.jpg')
team1_data = {
    'name'      : 'team1', 
    'wins'      : 7, 
    'loses'     : 0,
    'pic1_avg'  : np.mean(team1_pic),
    'pic1_std'  : np.std(team1_pic),
    'pic1_value': team1_pic.data,
    'pic1_ndim' : team1_pic.ndim, 
    'pic1_dim1' : team1_pic.shape[0],
    'pic1_dim2' : team1_pic.shape[1],
    'pic1_dim3' : team1_pic.shape[2],
    'pic1_dim4' : -9999,
    'pic1_dtype': team1_pic.dtype.name
}
team1_data
```

```python
team2_pic  = plt.imread('images/team2.jpg')
team2_data = {
    'name'      : 'team2', 
    'wins'      : 2, 
    'loses'     : 2,
    'pic1_avg'  : np.mean(team2_pic),
    'pic1_std'  : np.std(team2_pic),
    'pic1_value': team2_pic.data,
    'pic1_ndim' : team2_pic.ndim, 
    'pic1_dim1' : team2_pic.shape[0],
    'pic1_dim2' : team2_pic.shape[1],
    'pic1_dim3' : team2_pic.shape[2],
    'pic1_dim4' : -9999,
    'pic1_dtype': team1_pic.dtype.name
}
team2_data
```

```python
teams_data = [team1_data, team2_data]
teams_data
```

```python
teams_insert_sql = """
insert into teams
(
    id,
    name,
    wins,
    loses,
    pic1
)
values
(
    null,
    :name,
    :wins,
    :loses,
    ArrayInfo
    (
        :pic1_avg,
        :pic1_std,
        :pic1_ndim,
        :pic1_dim1,
        :pic1_dim2,
        :pic1_dim3,
        :pic1_dim4,
        :pic1_dtype,
        :pic1_value
    )
)
"""

print(teams_insert_sql)
```

```python
try:
    # connect
    conn = odb.connect(user, pswd, sid)
    
    # insert player objects
    cursor = conn.cursor()
    cursor.setinputsizes(
        name=odb.STRING,
        wins=odb.NUMBER,
        loses=odb.NUMBER,
        pic1_avg=odb.NUMBER,
        pic1_std=odb.NUMBER,
        pic1_ndim=odb.NUMBER,
        pic1_dim1=odb.NUMBER,
        pic1_dim2=odb.NUMBER,
        pic1_dim3=odb.NUMBER,
        pic1_dim4=odb.NUMBER,
        pic1_dtype=odb.STRING,
        pic1_value=odb.BLOB,
    )
    cursor.executemany(teams_insert_sql, teams_data)
    
    # done
    conn.commit()
    print("filled Table players succesfully")
  
except odb.DatabaseError, exc:
    print(exc)

finally:
    cursor.close()
    conn.close()
```

```python
try:
    conn = odb.connect(user, pswd, sid)
    teams_df = psql.read_sql('select * from teams', conn)

except odb.DatabaseError, exc:
    print(exc)

finally:
    conn.close()
```

```python
def get_player_data(team, name, goals, starts, image1, image2, image3):
    player_data = {
        'team'        : team,
        'name'        : name,
        'goals'       : goals, 
        'starts'      : starts,
        'image1_avg'  : np.mean(image1),
        'image1_std'  : np.std(image1),
        'image1_value': image1.data,
        'image1_ndim' : image1.ndim, 
        'image1_dim1' : image1.shape[0],
        'image1_dim2' : image1.shape[1],
        'image1_dim3' : image1.shape[2],
        'image1_dim4' : -9999,
        'image1_dtype': image1.dtype.name,
        'image2_avg'  : np.mean(image2),
        'image2_std'  : np.std(image2),
        'image2_value': image2.data,
        'image2_ndim' : image2.ndim, 
        'image2_dim1' : image2.shape[0],
        'image2_dim2' : image2.shape[1],
        'image2_dim3' : image2.shape[2],
        'image2_dim4' : -9999,
        'image2_dtype': image2.dtype.name,
        'image3_avg'  : np.mean(image3),
        'image3_std'  : np.std(image3),
        'image3_value': image3.data,
        'image3_ndim' : image3.ndim, 
        'image3_dim1' : image3.shape[0],
        'image3_dim2' : image3.shape[1],
        'image3_dim3' : image3.shape[2],
        'image3_dim4' : -9999,
        'image3_dtype': image3.dtype.name,
    }
    return player_data
```

```python
players_insert_sql = """
insert into players
(
    id,
    name,
    goals,      
    starts,     
    image1,
    image2, 
    image3,
    teams_id
)
values
(
    null,
    :name,
    :goals,
    :starts,
    ArrayInfo(:image1_avg, :image1_std, :image1_ndim, :image1_dim1, :image1_dim2, :image1_dim3, :image1_dim4, :image1_dtype, :image1_value),
    ArrayInfo(:image2_avg, :image2_std, :image2_ndim, :image2_dim1, :image2_dim2, :image2_dim3, :image2_dim4, :image2_dtype, :image2_value),
    ArrayInfo(:image3_avg, :image3_std, :image3_ndim, :image3_dim1, :image3_dim2, :image3_dim3, :image3_dim4, :image3_dtype, :image3_value),
    (select teams.id from teams where teams.name=:team)
)
"""

print(players_insert_sql)
```

```python
try:
    # connect
    conn = odb.connect(user, pswd, sid)
    
    # insert player objects
    cursor = conn.cursor()
    cursor.setinputsizes(
        image1_avg   = odb.NUMBER,
        image1_std   = odb.NUMBER,
        image1_ndim  = odb.NUMBER,
        image1_dim1  = odb.NUMBER,
        image1_dim2  = odb.NUMBER,
        image1_dim3  = odb.NUMBER,
        image1_dim4  = odb.NUMBER,
        image1_dtype = odb.STRING,
        image1_value = odb.BLOB,
        image2_avg   = odb.NUMBER,
        image2_std   = odb.NUMBER,
        image2_dim1  = odb.NUMBER,
        image2_dim2  = odb.NUMBER,
        image2_dim3  = odb.NUMBER,
        image2_dim4  = odb.NUMBER,
        image2_dtype = odb.STRING,
        image2_ndim  = odb.NUMBER,
        image2_value = odb.BLOB,
        image3_avg   = odb.NUMBER,
        image3_std   = odb.NUMBER,
        image3_dim1  = odb.NUMBER,
        image3_dim2  = odb.NUMBER,
        image3_dim3  = odb.NUMBER,
        image3_dim4  = odb.NUMBER,
        image3_dtype = odb.STRING,
        image3_ndim  = odb.NUMBER,
        image3_value = odb.BLOB,
        team         = odb.STRING,
        name         = odb.STRING,
        goals        = odb.NUMBER,
        starts       = odb.NUMBER
    )
    cursor.executemany(players_insert_sql, players_data)
    
    # done
    conn.commit()
    print("filled Table players succesfully")
  
except odb.DatabaseError, exc:
    print(exc)

finally:
    cursor.close()
    conn.close()
```
