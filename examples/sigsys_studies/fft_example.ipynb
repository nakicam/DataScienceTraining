## De noising tutorial

###### Radar example

```python
T = 30.;   # time window
N = 512;  # number of fourier modes,
t2 = np.linspace(-T,T, N+1)  # N+1 "samples"
t  = t2[:-1]
delta_t = 2*T/(N+1)

# angular frequency
ws = 2*np.pi/(delta_t)
w  = ws/N*np.concatenate([np.arange(0,N/2,1), np.arange(-N/2,0,1)])
```

```python
# Nyquist freq, i.e. max freq
w_max = N*ws/2
w_max
```

```python
# using numpy's built in function
w_np= 2*np.pi*fftfreq(N,delta_t)
np.sum(w_np-w)
```

```python
# true signal
y = 1.0/np.cosh(t)

plt.plot(t,y)
plt.xlabel('signal', fontsize = 20)
plt.xlabel('time', fontsize = 20)
plt.ylabel('response', fontsize=20)
```

```python
plt.plot(fftshift(w), np.abs(fftshift(y_fft)))
plt.ylabel(r'$|\tilde{y}(\omega)|$', fontsize = 30)
plt.xlabel(r'$\omega$ [rad/sec]', fontsize = 30)
plt.title('FFT', fontsize=20)
```

###### Add noise

```python
noise=10;
y_fft = fft(y); 

y_fft_wNoise = y_fft + noise*(np.random.normal(0,1.0, N)+ 1.0j*np.random.normal(0,1.0, N))

y_noise = ifft(y_fft_wNoise)
```

```python
y_noise.shape
```

```python
plt.plot(t, y_noise)
plt.ylabel(r'y(t)', fontsize = 30)
plt.xlabel(r'$t$ [sec]', fontsize = 30)
plt.title('time domain', fontsize=20)
```

```python
plt.plot(fftshift(w), np.abs(fftshift(y_fft_wNoise)))
plt.ylabel(r'$|\tilde{y}(\omega)|$', fontsize = 30)
plt.xlabel(r'$\omega$ [rad/sec]', fontsize = 30)
plt.title('FFT', fontsize=20)
```

We now apply a gaussian filter, in frequency domain:
$$
\mathcal{F}(\omega) = \exp( - \tau(\omega-\omega_0))
$$
here $\omega_0 = 0$ since we are dealing with a hyperpolic secant function.  Also, recall that convolution in the time domain correspsonds to multiplication in the frequency domain, and so we multiply the signal by the fiter's FFT.

```python
gauss_filter = np.exp(-0.2*(w)**2)
y_fft_filtered = gauss_filter*y_fft_wNoise
y__filtered = ifft(y_fft_filtered)
```

```python
plt.plot(fftshift(w), np.abs(fftshift(y_fft_filtered)))
plt.ylabel(r'$|\tilde{y}(\omega)|$', fontsize = 30)
plt.xlabel(r'$\omega$ [rad/sec]', fontsize = 30)
plt.title('FFT (filtered)', fontsize=20)
```

```python
plt.plot(t, y__filtered)
plt.ylabel(r'y(t)', fontsize = 30)
plt.xlabel(r'$t$ [sec]', fontsize = 30)
plt.title('time domain (filtered)', fontsize=20)
```

## apply filter to NBC data

```python
# target universe size
univ = 115e3

# compute ratings in GRPs
y = (100*data[['average_audience']]/univ).values  # [GRP]

# select training set
y = y[0:638]

# generate time array
N = len(y)
delta_t = 1/365.  # inverse sampling freq [year]
tf = (N-1)*delta_t
ti = 0
t = np.linspace(0, (N-1)*delta_t, N)

# compute FFT and plot
y_fft = fft(y)
f = fftfreq(N, d=delta_t)

# shift for display
f_plot = fftshift(f)
y_plot = fftshift(y_fft)

plt.plot(f_plot, (1.0/N * np.abs(y_plot)))
plt.xlabel('freq [cycles/year]', fontsize = 20)
plt.ylabel('amplitude', fontsize=20)
plt.grid()
plt.show()
```

```python
y_fft.shape
```

```python
alpha = 5
# f0_list = [7,31,52, -7,-31,-52]
f0_list = np.arange(-187,187,1)
gauss_filter = 0.0
for f0 in f0_list :
    gauss_filter = gauss_filter+ np.exp(-alpha*(f-f0)**2)/np.sqrt(2*np.pi)
    
gauss_filter = gauss_filter.reshape([N,1])    
y_fft_filtered = gauss_filter*y_fft
y_filtered = ifft(y_fft_filtered)
```

```python
y_fft_filtered.shape
```

```python
plt.plot(fftshift(f), (1.0/N * np.abs(fftshift(y_fft_filtered))))
plt.xlabel('freq [cycles/year]', fontsize = 20)
plt.ylabel('amplitude', fontsize=20)
plt.grid()
plt.show()
```
