## Overview

A continuous-time signal can be processed by processing its samples through a discrete-time system. For this purpose, it is important to maintain the signal sampling rate high enough to permit the reconstruction of the original signal from these samples without error (or with an error within a given tolerance). The necessary quantitative framework for this purpose is provided by the sampling theorem derived in Section 8.1.

The sampling theory is the bridge between the continuous-time and the discrete-time worlds. The information inherent in a sampled continuous-time signal is equivalent to that of a discrete-time signal. A sampled continuous-time signal is a sequence of impulses, while a discrete-time signal presents the same information as a sequence of numbers. These are basically two different ways of presenting the same data. Clearly, all the concepts in the analysis of sampled signals apply to discrete-time signals. We should not be surprised to
see that the Fourier spectra of the two kinds of signal are also the same (within a multiplicative constant).

## 3.1: The Sampling Theorem

We now show that a real signal whose spectrum is bandlimited to $B$ Hz $[\Xw= 0$ for $|\w| > 2\pi B]$ can be reconstructed exactly (without any error) from its samples taken uniformly at a rate $\fs > 2B$ samples per second. In other words, the minimum sampling frequency is $\fs = 2B$ Hz.

To prove the sampling theorem, consider a signal $\xt$ (Fig. 8.1a) whose spectrum is band-limited to $B$ Hz (Fig. 8.1b). For convenience, spectra are shown as functions of $\w$ as well as of $f$ (hertz). Sampling $\xt$ at a rate of $\fs$ Hz ($\fs$ samples per second) can be accomplished by multiplying $\xt$ by an impulse train $\deltaT(t)$ (Fig. 8.1c), consisting of unit impulses repeating periodically every $T$ seconds, where $T = 1/\fs$. The schematic of a sampler is shown in Fig. 8.1d. The resulting sampled signal $\xt$ is shown in Fig. 8.1e.

The sampled signal consists of impulses spaced every $T$ seconds (the sampling interval). The $n^{th}$ impulse, located at $t = nT$,
has a strength $x(nT)$, the value of $\xt$ at $t = nT$.

\begin{align}
\xbt = \xt\deltaT(t) = \sum_n x(nT) \delta(t - nT).
\end{align}

Because the impulse train $\deltaT$ is a periodic signal of period $T$, it can be expressed as a trigonometric Fourier series like that already obtained in Example 6.7 [Eq. (6.39)]

\begin{align}
\deltaT(t) = \frac{1}{T}\left[1 + 2 \cos \ws t + 2\cos 2\ws t + 2\cos 3\ws t + \dots \right]
\end{align}

Therefore

\begin{align}
\xbt = \xt\deltaT(t) = \frac{1}{T}\left[\xt + 2\xt \cos \ws t + 2\xt \cos 2\ws t + 2\xt \cos 3\ws t + \dots \right]
\end{align}

To find $\Xbw$, the Fourier transform of $\xbt$, we take the Fourier transform of the right-hand side of Eq. (8.3), term by term. The transform of the first term in the brackets is $\Xw$. The transform of the second term $2\xt \cos \ws t$ is $X(\w - \ws) + X (\w + \ws)$ [see Eq. (7.41)]. This represents spectrum $\Xw$ shifted to $\ws$ and $-\ws$. Similarly, the transform of the third term $2\xt \cos 2\ws t$ is $X(\w - 2\ws) + X(\w + 2\ws)$, which represents the spectrum $\Xw$ shifted to $2\ws$ and $-2\ws$, and so on to infinity. This result means that the spectrum $\Xbw$ consists of $\Xw$ repeating periodically with period $\ws = 2\pi/T$ rad/s, or $\fs = 1/T$ Hz, as depicted in Fig. 8.1f. There is also a constant multiplier $1/T$ in Eq. (8.3). Therefore

\begin{align}
\Xbw = \frac{1}{T}\sum_{n=-\infty}^{+\infty} X(\w - n\ws).
\end{align}

If we are to reconstruct $\xt$ from $\xbt$, we should be able to recover $\Xw$ from $\Xbw$. This recovery is possible if there is no overlap between successive cycles of $\Xw$. Figure 8.1f indicates that this requires

$$\fs > 2B.$$

Also, the sampling interval $T = 1/\fs$. Therefore

$$T < \frac{1}{2B}.$$

Thus, as long as the sampling frequency $\fs$ is greater than twice the signal bandwidth $B$ (in hertz), $\Xbw$ consists of nonoverlapping repetitions of $\Xw$. Figure 8.1f shows that the gap between the two adjacent spectral repetitions is $\fs - 2B$ Hz, and $\xt$ can be recovered from its samples $\xbt$ by passing the sampled signal $\xbt$ through an ideal lowpass filter having a bandwidth of any value between $B$and $\fs - B$ Hz. The minimum sampling rate $\fs = 2B$ required to recover $\xt$ from its samples $\xbt$ is called the Nyquist rate for $\xt$, and the corresponding sampling interval $T= 1/2B$ is called the Nyquist interval for $\xt$. Samples of a signal taken at its Nyquist rate are the Nyquist samples of that signal.

**Note 1:** 
We are saying that the Nyquist rate $2B$ Hz is the minimum sampling rate required to preserve the information of $\xt$. This contradicts Eq. (8.5), where we showed that to preserve the information of $\xt$, the sampling rate $\fs$ needs to be greater than $2B$ Hz. Strictly speaking, Eq. (8.5) is the correct statement. However, if the spectrum $\Xw$ contains no impulse or its derivatives at the highest frequency $B$ Hz, then the minimum sampling rate $2B$ Hz is adequate. In practice, it is rare to observe $\Xw$ with an impulse or its derivatives at the highest frequency. If the contrary situation were to occur, we should use Eq. (8.5).

**Note 2:**
The sampling theorem proved here uses samples taken at uniform intervals. This condition is not necessary. Samples can be taken arbitrarily at any instants as long as the sampling instants are recorded and there are, on average, $2B$ samples per second.[2] The essence of the sampling theorem was known to mathematicians for a long time in the form of the interpolation formula [see later, Eq. (8.11)]. The origin of the sampling theorem was attributed by H. S. Black to Cauchy in 1841. The essential idea of the sampling theorem was rediscovered in the 1920s by Carson, Nyquist, and Hartley.

#### example 8.1

In this example, we examine the effects of sampling a signal at the Nyquist rate, below the Nyquist rate (undersampling), and above the Nyquist rate (oversampling). Consider a signal $x(t)= \sin 2(5\pi t)$ (Fig. 8.2a) whose spectrum is $X(\w)= 0.2 \Delta (\w/20\pi)$ (Fig. 8.2b). The bandwidth of this signal is 5 Hz ($10\pi$ rad/s). Consequently, the Nyquist rate is 10 Hz; that is, we must sample the signal at a rate no less than 10 samples/s. The Nyquist interval is $T = 1/2B = 0.1$ second.

Recall that the sampled signal spectrum consists of $(1/T) \Xw = (0.2/T) \Delta (ω/20pi)$ repeating periodically with a period equal to the sampling frequency $\fs$ Hz. We present this information in Table 8.1 for three sampling rates: $\fs = 5$ Hz (undersampling), 10 Hz (Nyquist rate), and 20 Hz (oversampling).

\begin{array}{|l|l|l|l|}
\hline
\fs\ \text{Hz} & T\ \text{second} & \frac{1}{T}\Xw                       & \text{comments}        \\
\hline
5              & 0.2              & \Delta \left(\frac{\w}{20\pi}\right) & \text{undersampled}    \\
10             & 0.1              & 2\Delta \left(\frac{\w}{20\pi}\right)& \text{Nyquist sampled} \\
20             & 0.05             & 4\Delta \left(\frac{\w}{20\pi}\right)& \text{oversampled}     \\
\hline
\end{array}

In the first case (undersampling), the sampling rate is 5 Hz (5 samples/s), and the spectrum $(1/T)\Xw$ repeats every 5 Hz ($10\pi$ rad/s). The successive spectra overlap, as depicted in Fig. 8.2d, and the spectrum $\Xw$ is not recoverable from $\Xbw$; that is, $\xt$ cannot be reconstructed from its samples $\xbt$ in Fig. 8.2c. In the second case, we use the Nyquist sampling rate of 10 Hz (Fig. 8.2e).  The spectrum $\Xw$ consists of back-to-back, nonoverlapping repetitions of $(1/T)\Xw$ repeating every 10 Hz. Hence, $\Xw$ can be recovered from $\Xbw$ using an ideal lowpass filter of bandwidth 5 Hz (Fig. 8.2f). Finally, in the last case of oversampling (sampling rate 20 Hz), the spectrum $\Xbw$ consists of nonoverlapping repetitions of $(1/T)\Xw$ (repeating every 20 Hz) with empty bands between successive cycles (Fig. 8.2h). Hence, $\Xw$ can be recovered from $\Xbw$ by using an ideal lowpass filter or even a practical lowpass filter (shown dashed in Fig. 8.2h).

#### Exercise E8.1

Find the Nyquist rate and the Nyquist interval for the signals 
- a) $\xt = \sinc(100\pi t)$
- b) $\xt = \sinc(100\pi t) + \sinc (50 \pi t)$.

##### part a:

First find the FT of $\xt$ which uses the fact that:

$$\rect(t) \Longleftrightarrow \sinc(\w/2)$$

where $\rect(t)$ is the "rectangle" function centered at origin:

\begin{align}
\operatorname{rect}(x) = 
  \left\{
  \begin{aligned}
  0           \qquad |x| > \frac{1}{2} \\
  \frac{1}{2} \qquad |x| = \frac{1}{2} \\
  1           \qquad |x| < \frac{1}{2} \\
  \end{aligned}
  \right.
\end{align}

Using **duality**, 

$$F(t) \Longleftrightarrow \twopi f(-\omega),$$

we get 

$$\sinc(t/2) = \twopi \rect(-\w).$$

Using scaling property of FT with $a = 200\pi$, we get

\begin{align}
\Xw &= \frac{\twopi}{|200\pi|} \rect \left(\frac{-\w}{|200\pi|}\right) \\
    &= \frac{1}{100}\rect\left(\frac{f}{100}\right)
\end{align}

To get the bandwidth, we use the fact that 

$$\frac{f_{max}}{100} = \frac{1}{2}$$

at the boundary so,

$$B = f_{max} = 50\ \text{hz}.$$

The sampling rate $T = 2B = 100$ Hz and the sampling interval is $\fs = 1/T = 0.01$ seconds.

##### part b:

The FT of $\xt$ is:

\begin{align}
\Xw &= \frac{1}{100}\rect\left(\frac{f}{100}\right) + \frac{1}{50}\rect\left(\frac{f}{50}\right).
\end{align}

The bandwidth is determined by the first term so the sampling rate $T = 2B = 100$ Hz and the sampling interval is $\fs = 1/T = 0.01$ seconds.


### For skeptics only

Rare is a reader, who, at first encounter, is not skeptical of the sampling theorem. It seems impossible that Nyquist samples can define the one and the only signal that passes through those sample values. We can easily picture infinite number of signals passing through a given set of samples. However, among all these (infinite number of) signals, only one has the minimum bandwidth $B \leq 1/2T$ Hz, where $T$ is the sampling interval. See Prob. 8.2−9.

To summarize, for a given set of samples taken at a rate of $\fs$ Hz, there is only one signal of bandwidth $B \leq \fs/2$ that passes through those samples. All other signals that pass through those samples have bandwidth higher than $\fs/2$, and the samples are sub-Nyquist rate samples for those signals.
