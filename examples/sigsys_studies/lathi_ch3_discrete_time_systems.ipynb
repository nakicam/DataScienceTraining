## Discrete-Time Signal

We denote a discrete-time (DT) signal by $x[n]$ where $n \in \mathbb{Z}$.  These sorts of signals are basically a sequence of numbers.  Examples:

 * populations
 * national income models
 * radar tracking
 * **sampling** of continuous-time signals to be used in further processing such as a digital filter
 
This last point is where we are most concerned.  A discrete-time signal can be obtained from a continous-time (CT) signal $x(t)$ by sampling the function $n$-times.  If we do this a regulary spaced intervals $T$, we obtain a sampled function $x[n] = x(nT)$.

```python
def f(t):
    return np.exp(-t**2)

T   = 0.1
lim = [-5, 5]
N   = np.floor(np.abs(lim[1] - lim[0])/T)
n   = np.arange(-N/2, N/2)
x   = f(n*T)

plt.vlines(n, [0], x, linewidth=2, color='b')
plt.xlabel(r"$n \rightarrow$")
plt.xlim([-30,30])
plt.ylabel(r"$x[n] = x(nT)$")
plt.show()
```

Digital filters can process continuous-time signals by discrete-time systems using a Continous-to-Discrete converter (C/D) to sample the signal to discrete time, run it through the discrete-time system, and then converting it back with a Discrete-to-Continuous converter (D/C) to reconstruct the continuous signal.

### Size of a DT signal

#### Energy

The energy for a DT system is defined as

$$E_x = \suminf |x[n]|^2$$

Where we require $|x[n]| \rightarrow 0$ as $n \rightarrow \infty$.  If the energy exists, then $x$ is an *energy signal*.

#### Power

For non-energy signals, we can define the power for a DT system by taking a time average of the singal (if it exists).  The power is defined as

$$P_x = \lim_{N \rightarrow \infty}\frac{1}{2N+1}\sum_{-N}^{N} |x[n]|^2$$

For periodic signal, the time averaging need be performed over a single period.  If the power exists, then $x$ is an *power signal*.

As with CT signals, a DT signal can either be a energy signal or a power signal, but not both.  Also, a DT signal can be neither.
