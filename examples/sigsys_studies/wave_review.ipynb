## 1-D Wave Equation and Basic Properties

Deriving the wave equation is beyond the scope of this exercise.  It can be shown using classical mechanics that the general 1-D wave equation is

\begin{align}
    \frac{d^2\Psi}{dx^2} - \frac{1}{c^2}\frac{d^2\Psi}{dt^2} = 0
\end{align}

Solutions of the wave equation are given by a linear combination of sinusoids (or complex exponentials):

\begin{align}
    \Psi(x, t) &\sim \sum_{r} a_r \Psi_r(x, t) \\
           &\sim \sum_{r} a_r\ {\rm exp}\left\{ \pm i \frac{\omega_r}{v}(x \pm vt) \right\}
\end{align}

We need only consider one $\Psi_r(x, t)$ since the $a_r$ can absorb the $\pm$ in the form of a complex phase shift.  So without loss of generality, we drop the subscript $r$ for now and only consider solutions of the form:

\begin{align}
    \Psi(x, t) \sim {\rm exp}\left\{ - i \frac{\omega}{v}(x - vt) \right\}
\end{align}

It is customary to define the *progagation constant* or *wave number* $k$ (dimensions $[L^{-1}]$) by:

$$k^2 = \frac{\omega^2}{v^2}$$

Also, the wavelength $\lambda$ is the distance required to complete one vibration or cycle:

$$\lambda = \frac{v}{f} = \frac{2 \pi v}{\omega}$$

and can be related to $k$ by

$$k = \frac{2 \pi}{\lambda} = \frac{\omega}{v}$$

So, we can recast the expression for $\Psi(x,t)$ as

\begin{align}
    \Psi(x, t) &\sim e^{ -i k(x - vt)} \\
           &= e^{i(\omega t - k x)}
\end{align}

### plot of a single sinusoid:

For the purpose of this coding exercise, let's use 

$$\Psi(x,t) = A\ \cos(\omega t - k x)$$

We can get away with this since you can always write $$\cos(x) = \frac{e^{ix} + e^{-ix}}{2}$$.

Variables in code:
* x = spatial coord [m]
* t = time coord [s]
* f = freq of wave [hz]
* w = $\omega$ or angular freq of wave [rad/s]
* k = $k$ or wave number of wave [m$^{-1}$]
* A = amplitude of wave [m]

```python
A    = 2   # amplitude
f    = 0.5 # freq
t    = 0   # time
v    = 1   # m/s
w    = 2*np.pi*f
k    = w/v

x_fs = 100 # spatial sampling frequency
x0   = 0   # initial position
xf   = 10   # final final
dx   = np.abs(xf-x0)/x_fs
x    = np.arange(x0, xf, dx)

t_fs = 100 # time sampling freq
t0   = 0   # initial time
tf   = 10  # final time
dt   = np.abs(tf-t0)/t_fs
t    = np.arange(t0, tf, dt)

(T,X) = np.meshgrid(t,x)

# construct array of values
psi  = A*np.cos(w*T - k*X)
psi.shape
```

#### fixed position x

```python
x1_idx = 0
x2_idx = 7
plt.plot(t, psi[:,x1_idx], linewidth=2, color='red')
plt.plot(t, psi[:,x2_idx], linewidth=2, color='blue')
plt.xlabel('$t$ (s)')
plt.ylabel('$\Psi(x=fixed, t)$')
plt.axis([t0, tf,-1.5*A,1.5*A])
plt.tight_layout()
plt.legend(['x = 0', 'x = 7.7'], fontsize=16)
plt.legend([('x = %0.1f'%v) for v in x[[x1_idx,x2_idx]]])
plt.show()
```

#### fixed t

```python
plt.plot(x, psi[0,:], linewidth=2, color='red')
plt.plot(x, psi[1,:], linewidth=2, color='blue')
plt.plot(x, psi[2,:], linewidth=2, color='green')
plt.xlabel('$x$ (m)')
plt.ylabel('$\Psi(x, t=fixed)$')
plt.axis([t0, tf,-1.5*A,1.5*A])
plt.tight_layout()
plt.legend([('t = %0.1f'%v) for v in t[0:4]])
plt.show()
```

## Fourier Transform of single sinusoid

Now we want to consider the $\omega$--$k$ content of the sinusoid.  The is done by a Fourier Transform.  

### Fixed x $\rightarrow$ frequency content

First let's consider a Fourier Transform (FT) in time.  This for the continuous case case be done by:

\begin{align}
    \hat{\Psi}(x=x_{fixed}, \omega)
    &= \int_{-\infty}^{+\infty} dt\ \Psi(x=x_{fixed}, t)\ e^{-i\omega t} \\
    &= \int_{-\infty}^{+\infty} dt\ A \cos(\omega_0 t - kx_{fixed}) e^{-i\omega t} \\
    &= \int_{-\infty}^{+\infty} dt\ A \left\{\frac{e^{+i(\omega_0 t - kx_{fixed})} + e^{i-(\omega_0 t - kx_{fixed})}}{2}\right\} e^{-i\omega t} \\
    &= \frac{A}{2}\left\{ e^{-i k x_{fixed}}\int_{-\infty}^{+\infty} dt\ e^{+i t (\omega_0 - \omega)} + e^{+i k x_{fixed}}\int_{-\infty}^{+\infty} dt\ e^{-i t (\omega_0 + \omega)} \right\}
\end{align}

Recall that the covolution of a complex exponential yields a delta function ([link](http://en.wikipedia.org/wiki/Dirac_delta_function)).  Mainly,

\begin{align}
    \delta(x - a) = \frac{1}{2\pi} \int_{-\infty}^{+\infty} dp\ e^{+ip(x-a)}
\end{align}

and 

$$\delta(-x) = \delta(x)$$

Using this, we get:

\begin{align}
    \hat{\Psi}(x=x_{fixed}, \omega) = 
    \pi A \left\{e^{-i k x_{fixed}} \delta(\omega - \omega_0) + e^{+i k x_{fixed}}\delta(\omega + \omega_0)\right\}
\end{align}

### Power Spectral Density

The $|\hat{\Psi}(k)|$ can be plotted to give a sense of where the spectral components lie in the $k$ domain.  We do this for a fixed $t$.

\begin{align}
    |\hat{\Psi}(\omega)|^2 &= |\pi A \left\{e^{-i k x_{fixed}} \delta(\omega - \omega_0) + e^{+i k x_{fixed}}\delta(\omega + \omega_0)\right\}|^2 \\
    &= \pi^2 |A|^2 \left\{e^{-i k x_{fixed}} \delta(\omega - \omega_0) + e^{+i k x_{fixed}}\delta(\omega + \omega_0)\right\} 
    \times \left\{e^{+i k x_{fixed}} \delta(\omega - \omega_0) + e^{-i k x_{fixed}}\delta(\omega + \omega_0)\right\} \\
    &= \pi^2 |A|^2 \left\{\delta(\omega - \omega_0)^2 + \delta(\omega + \omega_0)^2 + 2\cos(kx_{fixed})\delta(\omega - \omega_0)\delta(\omega + \omega_0) \right\} \\
    &= \pi^2 |A|^2 \left\{\delta(\omega - \omega_0)^2 + \delta(\omega + \omega_0)^2 \right\} \\
    &= \pi^2 |A|^2 \left\{\delta(\omega - \omega_0) + \delta(\omega + \omega_0) \right\}
\end{align}

This claims that for a single sinusoid, their must be two "spikes" at $\pm$ sinusoid's frequency.  This is expected since we are dealing with two "components" of the expansion of $\Psi(x,t)$ in a basis of complex exponentials.

```python
import scipy.fftpack
import scipy.fftpack
x1_idx = 0
x2_idx = 7
psi_x1 = psi[:,x1_idx]
psi_x2 = psi[:,x2_idx]
N      = psi_x1.size
psi_x1_hat = np.around(scipy.fftpack.fft(psi_x1, N), decimals=5)
psi_x2_hat = np.around(scipy.fftpack.fft(psi_x2, N), decimals=5)
omega_hat  = scipy.fftpack.fftfreq(N, 1./t_fs)
psi_x1_hat.shape, psi_x2_hat.shape, omega_hat.shape
```

```python
plt.plot(omega_hat, np.abs(psi_x1_hat)**2, 'r-' , linewidth=2)
plt.plot(omega_hat, np.abs(psi_x2_hat)**2, 'b--', linewidth=2)
plt.xlabel('$\omega$ ($s^{-1}$)')
plt.ylabel('$\Psi(x=fixed, \omega)$')
plt.tight_layout()
plt.legend([('x = %0.1f'%v) for v in x[[x1_idx,x2_idx]]])
plt.show()
```

### Fixed t $\rightarrow$ wave number content

First let's consider a Fourier Transform (FT) in time.  This for the continuous case case be done by:

\begin{align}
    \hat{\Psi}(k, t_{fixed})
    &= \int_{-\infty}^{+\infty} dx\ \Psi(x=x, t_{fixed})\ e^{-i k x} \\
    &= \int_{-\infty}^{+\infty} dx\ A \cos(\omega t_fixed - k_0 x) e^{-i k x} \\
    &= \int_{-\infty}^{+\infty} dx\ A \left\{\frac{e^{+i(\omega t_{fixed} - k_0 x)} + e^{i-(\omega t_{fixed} - k_0 x)}}{2}\right\} e^{-ikx} \\
    &= \frac{A}{2}\left\{ e^{+i \omega t_{fixed}}\int_{-\infty}^{+\infty} dx\ e^{-i x (k_0 + k)} + e^{-i \omega t_{fixed}}\int_{-\infty}^{+\infty} dx\ e^{+i x (k_0 - k)} \right\}
\end{align}

Recall that the covolution of a complex exponential yields a delta function ([link](http://en.wikipedia.org/wiki/Dirac_delta_function)).  Mainly,

\begin{align}
    \delta(x - a) = \frac{1}{2\pi} \int_{-\infty}^{+\infty} dp\ e^{+ip(x-a)}
\end{align}

and 

$$\delta(-x) = \delta(x)$$

Using this, we get:

\begin{align}
    \hat{\Psi}(k, t_{fixed})
    &= \pi A \left\{e^{+i \omega t_{fixed}} \delta(k_0 + k) + e^{-i \omega t_{fixed}}\delta(k_0 - k)\right\}
\end{align}

### Power Spectral Density

The $|\hat{\Psi}(k)|$ can be plotted to give a sense of where the spectral components lie in the $k$ domain.  We do this for a fixed $t$.

\begin{align}
    |\hat{\Psi}(k)|^2 \\
    &= |\pi A \left\{e^{+i \omega t_{fixed}} \delta(k_0 + k) + e^{-i \omega t_{fixed}}\delta(k_0 - k)\right\}|^2 \\
    &= \pi^2 |A|^2 \left\{e^{+i \omega t_{fixed}} \delta(k_0 + k) + e^{-i \omega t_{fixed}}\delta(k_0 - k)\right\}
    \times \left\{e^{-i \omega t_{fixed}} \delta(k_0 + k) + e^{+i \omega t_{fixed}}\delta(k_0 - k)\right\} \\
    &= \pi^2 |A|^2 \left\{\delta(k_0 + k)^2 + \delta(k_0 - k)^2 + 2\cos(\omega t_{fixed})\delta(k_0 + k)\delta(k_0 - k) \right\} \\
    &= \pi^2 |A|^2 \left\{\delta(k_0 + k)^2 + \delta(k_0 - k)^2 \right\} \\
    &= \pi^2 |A|^2 \left\{\delta(k_0 + k)   + \delta(k_0 - k) \right\}
\end{align}

This claims that for a single sinusoid, their must be two "spikes" at $\pm$ sinusoid's wave number.  Again, this is expected since we are dealing with two "components" of the expansion of $\Psi(x,t)$.

```python
import scipy.fftpack
t1_idx = 0
t2_idx = 7
psi_t1 = psi[t1_idx,:]
psi_t2 = psi[t2_idx,:]
N      = psi_t1.size
psi_t1_hat = np.around(scipy.fftpack.fft(psi_t1, N), decimals=5)
psi_t2_hat = np.around(scipy.fftpack.fft(psi_t2, N), decimals=5)
k_hat      = scipy.fftpack.fftfreq(N, 1./x_fs)
psi_t1_hat.shape, psi_t2_hat.shape, k_hat.shape
```

```python
plt.plot(k_hat, np.abs(psi_t1_hat), 'r-' , linewidth=2)
plt.plot(k_hat, np.abs(psi_t2_hat), 'b--', linewidth=2)
plt.xlabel('$k$ ($m^{-1}$)')
plt.ylabel('$\Psi(x=fixed, t)$')
plt.tight_layout()
plt.legend([('x = %0.1f'%v) for v in x[[x1_idx,x2_idx]]])
plt.show()
```
