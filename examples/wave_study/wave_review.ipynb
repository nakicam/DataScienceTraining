## 1-D Wave Equation and Basic Properties

Deriving the wave equation is beyond the scope of this exercise.  It can be shown using classical mechanics that the general 1-D wave equation is

\begin{align}
    \frac{d^2\Psi}{dx^2} - \frac{1}{c^2}\frac{d^2\Psi}{dt^2} = 0
\end{align}

Solutions of the wave equation are given by a linear combination of sinusoids (or complex exponentials):

\begin{align}
    \Psi(x, t) &\sim \sum_{r} a_r \Psi_r(x, t) \\
           &\sim \sum_{r} a_r\ {\rm exp}\left\{ \pm i \frac{\omega_r}{v}(x \pm vt) \right\}
\end{align}

We need only consider one $\Psi_r(x, t)$ since the $a_r$ can absorb the $\pm$ in the form of a complex phase shift.  So without loss of generality, we drop the subscript $r$ for now and only consider solutions of the form:

\begin{align}
    \Psi(x, t) \sim {\rm exp}\left\{ - i \frac{\omega}{v}(x - vt) \right\}
\end{align}

It is customary to define the *progagation constant* or *wave number* $k$ (dimensions $[L^{-1}]$) by:

$$k^2 = \frac{\omega^2}{v^2}$$

Also, the wavelength $\lambda$ is the distance required to complete one vibration or cycle:

$$\lambda = \frac{v}{f} = \frac{2 \pi v}{\omega}$$

and can be related to $k$ by

$$k = \frac{2 \pi}{\lambda} = \frac{\omega}{v}$$

So, we can recast the expression for $\Psi(x,t)$ as

\begin{align}
    \Psi(x, t) &\sim e^{ -i k(x - vt)} \\
           &= e^{i(\omega t - k x)}
\end{align}

### plot of a single sinusoid:

For the purpose of this coding exercise, let's use 

$$\Psi(x,t) = A\ cos(\omega t - k x)$$

We can get away with this since you can always write $$cos(x) = \frac{e^{ix} + e^{-ix}}{2}$$.

Variables in code:
* x = spatial coord [m]
* t = time coord [s]
* f = freq of wave [hz]
* w = $\omega$ or angular freq of wave [rad/s]
* k = $k$ or wave number of wave [m$^{-1}$]
* A = amplitude of wave [m]

#### fixed position x

```python
A   = 1   # amplitude
f   = 2   # freq
x   = 0   # x position
v   = 1   # m/s
fs  = 200 # sampling frequency
t0  = 0   # intial time
tf  = 2   # final time
phi = np.pi*0.25

dt   = np.abs(t0-tf)/fs
w    = 2*np.pi*f
k    = w/v
t    = np.arange(t0, tf+dt, dt)
psi1 = A*np.cos(w*t - k*x)

#change x to 7
x = 7.7
psi2 = A*np.cos(w*t - k*x)
```

## Fourier Transform of single sinusoid

Now we want to consider the $\omega$--$k$ content of the sinusoid.  The is done by a Fourier Transform.  

### Fixed x $\rightarrow$ frequency content

First let's consider a Fourier Transform (FT) in time.  This for the continuous case case be done by:

\begin{align}
    \hat{\Psi}(x=x_{fixed}, \omega)
    &= \int_{-\infty}^{+\infty} dt\ \Psi(x=x_{fixed}, t)\ e^{-i\omega t} \\
    &= \int_{-\infty}^{+\infty} dt\ A e^{i(\omega_0 t - kx_{fixed})} e^{-i\omega t} \\
    &= A e^{-i k x_{fixed}}\int_{-\infty}^{+\infty} dt\ e^{+i t (\omega_0 - \omega)}
\end{align}

Recall that the covolution of a complex exponential yields a delta function ([link](http://en.wikipedia.org/wiki/Dirac_delta_function)).  Mainly,

\begin{align}
    \delta(x - a) = \frac{1}{2\pi} \int_{-\infty}^{+\infty} dp\ e^{+ip(x-a)}
\end{align}

and 

$$\delta(-x) = \delta(x)$$

Using this, we get:

\begin{align}
    \hat{\Psi}(x=x_{fixed}, \omega) = 2 \pi A e^{-i k x_{fixed}}\delta(\omega - \omega_0)
\end{align}

For the coding example, using *cosine*, it actually works that we get two $\delta$ functions, one for each expontial that makes up the cosine:

\begin{align}
    \hat{\Psi}(x=x_{fixed}, \omega) = 
    \pi A e^{-i k x_{fixed}}\left\{\delta(\omega - \omega_0) + \delta(\omega + \omega_0)\right\}
\end{align}

### Power Spectral Density

The $|\hat{\Psi(\omega)}|$ can be plotted to give a sense of where the spectral components lie in the frequency domain.  We do this for a fixed $x = 0$.

\begin{align}
    |\hat{\Psi}(\omega)| &= |2 \pi A e^{-i k x_{fixed}}\delta(\omega - \omega_0)| \\
    &= 2 \pi |A| \delta(\omega - \omega_0)
\end{align}

This claims that for a single sinusoid, the amplitude must be a "spike" at the desired frequency.  This is expected since we are dealing with one "component" of the expansion of $\Psi(x,t)$ in a basis of complex exponentials.

For the coding example using *cosines*, we get two spikes:

\begin{align}
    |\hat{\Psi}(\omega)| = \pi |A| \left\{\delta(\omega - \omega_0) + \delta(\omega + \omega_0)\right\}
\end{align}


```python
psi_x0 = psi[x==x0].ravel()
N      = psi_x0.size
psi_x0.shape, N
```

```python
psi_x1 = psi[x==x[1]].ravel()
psi_x1.shape, N
```

```python
plt.plot(t, psi_x0, linewidth=2, color='red')
plt.plot(t, psi_x1, linewidth=2, color='blue')
plt.xlabel('time (s)')
plt.ylabel('$\Psi(x=fixed, t)$')
plt.axis([t0, tf,-1.5,1.5])
plt.tight_layout()
plt.legend(['x = %0.2f' % v for v in x[0:2]], fontsize=16)
plt.show()
```

```python
import scipy.fftpack

psi_x0_hat = np.around(scipy.fftpack.fft(psi_x0, N), decimals=5)
psi_x1_hat = np.around(scipy.fftpack.fft(psi_x1, N), decimals=5)
omega      = scipy.fftpack.fftfreq(N, 1./f)
psi_x0_hat.shape, psi_x1_hat.shape, omega.shape
```

```python
plt.plot(omega, np.abs(psi_x0_hat), 'r-' , linewidth=2)
plt.plot(omega, np.abs(psi_x1_hat), 'b--', linewidth=2)
plt.xlabel('time (s)')
plt.ylabel('$\Psi(x=fixed, t)$')
plt.tight_layout()
plt.legend(['x = %0.2f' % v for v in x[0:2]], fontsize=16)
plt.show()
```
